```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
```

```python
df = fetch_california_housing(as_frame = True).data.sample(500)
```

```python
df.head(3)
```

```python
plt.figure(figsize = (5.2, 5.2))
ax = df.Population.plot.box(vert = False, xlabel = "Population")
ax.set_yticklabels([])
plt.show()
```
